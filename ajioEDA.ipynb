checking duplicated values

```python
orders.duplicated().sum()
transactions.duplicated().sum()
customers.duplicated().sum()
products.duplicated().sum()
returns.duplicated().sum()
```

droping duplicated values

```python
orders=orders.drop_duplicates()
```

finding null values

```python
orders.isnull().sum()
```

```python
#if null value
#if numeric fill with median
#if textual value fill with mode or ffill or bfill

```

```python
orders['Qty']=orders['Qty'].fillna(orders['Qty'].median())
```

```python
customers.head(2)
```

joining datasets

```python
merged_cust_ord=merge(right=customers,left=orders,how='inner',on='C_ID')
```

```python
merged_cust_ord.head()
```

```python
gb=merged_cust_ord.groupby('State').agg({"Or_ID":'count'})
gb
```

```python
gb=gb.sort_values(by='Or_ID',ascending=False)
```

Sales and Revenue Analysis

orders by state

```python
sns.barplot(x=gb.index,y='Or_ID',data=gb,hue=gb.index)
plt.title('NO of orders in each state')
plt.ylabel('number of orders')
plt.xticks(rotation=90)
plt.show()
```

```python
orders['Order_Date']=to_datetime(orders['Order_Date'])
```

```python
orders=orders.merge(products,on='P_ID')
```

```python
orders.head()
```

```python
orders['Revenue']=(orders['Price']-orders['Discount'])*orders['Qty']
```

```python
orders.head(2)
```

monthly revenue

```python
monthly_revenue = orders.groupby(orders['Order_Date'].dt.to_period('M'))['Revenue'].sum().reset_index()
```

```python
monthly_revenue.head()
```

```python
sns.barplot(x='Order_Date',y='Revenue',data=monthly_revenue,hue='Order_Date')
plt.title('Monthly Revenue')
plt.ylabel('Revenue(in millions)')
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.show()
```

```python
full_data=orders.merge(transactions,on='Or_ID')
```

```python
full_data=full_data.merge(returns,on='Or_ID')
```

```python
full_data=full_data.merge(customers,on='C_ID')
```

```python
full_data.head()
```

```python
full_data.columns
```

customer age distribution

```python
full_data['Age'].plot(kind='hist',bins=20,rwidth=0.8, title='Customer Age Distribution',)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

top 10 customers

```python
top_customers = full_data.groupby('C_Name')['Revenue'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
top_customers
```

```python
sns.barplot(x='Revenue',y='C_Name',data=top_customers,hue='C_Name')
plt.title('Top 10 Customers by Revenue')
plt.ylabel('Customer Name')
plt.xlabel('Revenue')
plt.show()
```

top products

```python
top_product=full_data['P_Name'].value_counts().head(10)
```

```python
top_product
```

```python
plt.pie(top_product,labels=top_product.index,autopct='%1.1f%%',explode=[0.1]*len(top_product))
plt.title('Top 10 Products')
plt.show()
```

sales category

```python
category_sales = full_data.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
```

```python
category_sales
```

```python
plt.pie(category_sales,labels=category_sales.index,autopct='%1.1f%%',explode=[0.1]*len(category_sales))
plt.title('Sales by Category')
plt.show()
```
